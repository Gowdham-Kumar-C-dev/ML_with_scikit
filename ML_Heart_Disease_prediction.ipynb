```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('synthetic_heart_disease_dataset.csv')
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna(df['Alcohol_Intake'],inplace=True)
```

```python
df['Gender'].replace({'Male':1,'Female':0},inplace=True)
```

```python
df
```

```python
df['Alcohol_Intake'].replace({'Low':0,'Moderate':1,'High':2},inplace=True)
```

```python
df['Smoking'].replace({'Current':0,'Former':1,'Never':2},inplace=True)
```

```python
df['Diet'].replace({'Unhealthy':0,'Average':1,'Healthy':2},inplace=True)
```

```python
df['Physical_Activity'].replace({'Sedentary':0,'Moderate':1,'Active':2},inplace=True)
```

```python
df
```

```python
df['Stress_Level'].replace({'Low':0,'Medium':1,'High':2},inplace=True)
```

```python
df
```

```python
df.info()
```

```python
x=df[['Age','BMI','Smoking','Alcohol_Intake','Physical_Activity','Diet','Stress_Level','Previous_Heart_Attack','Cholesterol_Total']]
```

```python
y=df['Heart_Disease']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=DecisionTreeClassifier(random_state=42)
model.fit(x_train,y_train)
```

```python
y_pre=model.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pre)
```

```python
print(acc)
```
